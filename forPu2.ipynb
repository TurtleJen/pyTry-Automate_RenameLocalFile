# 0. 準備工作

```python
import openpyxl,os

wb = openpyxl.load_workbook('docs\s002.xlsx')
club = {}        # 社團名稱字典（社團名vs.sheet座標）
pic = {}         # 檔名字典（檔名vs.sheet座標）
loc = {}         # 攤位位置字典（走道名vs.欄）

sh2 = wb.get_sheet_by_name('攤位號碼(47後不得再修改)')
sh3 = wb.get_sheet_by_name('攤宣圖編碼')

path = r'C:\Users\user\PyTry\Automate_RenameLocalFile\items\攤宣圖集read'

# 走道名 與 欄位 對應，fill loc{}
def fill_loc(rowNumber):
    for loc_name in sh2[rowNumber]:
        if loc_name.value != None :
            loc[loc_name.column] = loc_name.value

# 將儲存格座標 轉成 攤位位置（走道名＋編號）
def coo_to_loc(coordinate):
    location = loc[coordinate[0]] + str( int(coordinate[1:]) - 1)
    return location
```

# 1. 讀社團名(縱向)

```python
def fill_club(start_column,end_column):    # start_column,end_column must be string
    start = ord(start_column)              # get ASCII code from column alphabet string
    end = ord(end_column)
    
    for alp in range(start,end):           # goal column zone 
        col = chr(alp)                     # get column alphabet string from ASCII code
    
        for sh2_i in sh2[col]:
            if (sh2_i.row != 1) and (sh2_i.value != None):
                club[sh2_i.value] =sh2_i.coordinate
```

# 2. 讀file name

```python
def fill_pic(path):
    picNum = 1    
    for file in os.listdir(path):
        pic[picNum] = file  
        picNum = picNum + 1 
```

# 3. 拿社團名去對file name，對中填表格 & 改檔名
------
* 拿社團名去對file name，對不到pass，對到：
  * 填sheet3 ，計數器數字、攤位編號、攤位名稱、原檔名
  * 改file name & 計數器加1

```python
# main
fill_loc(1)
fill_club('B','N')
fill_pic(path)

# 3. 拿社團名去對file name
count = 1
for clubName, clubCoo in club.items():                 # 從club字典拿到社團名，只取3個字來核對
    check = clubName[0:3]
    
    for picName in pic.values():                       # 從pic字典拿到圖檔檔名，只取3個字來核對        
        to_check = picName[4:7]
        
        if check != to_check : pass                    # 不處理對不中的狀況
        else :
            write_loc = coo_to_loc(clubCoo)            # 對中的社團儲存格座標轉成攤位編號
            newFileName = str(count).zfill(3)+'.jpg'   # 新圖檔名要3碼數字，須補0，還要加上副檔名
            
            a = 'A'+ str(count+1)                      # sheet3 A欄=圖片編號
            sh3[a] = newFileName[:-4]
            b = 'B'+ str(count+1)                      # sheet3 B欄=攤位編號
            sh3[b] = write_loc
            c = 'C'+ str(count+1)                      # sheet3 C欄=社團名稱
            sh3[c] = clubName
            e = 'E'+ str(count+1)                      # sheet3 E欄=原始圖片名稱
            sh3[e] = picName
            
            os.rename(os.path.join(path, picName), os.path.join(path, newFileName))
            
            count = count + 1  
```

# 4. 表格另存新檔
